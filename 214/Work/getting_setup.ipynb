I will assume that you have a basic working knowledge of the following equations and concepts:

Scale Factor: $a= 1/(1+z)$
* The scale factor in the FRW metric relates coordiantes $(r,\theta,\phi)$ to true physical distances.
* Proper distance = scale factor x comoving distance. $r=a(t) \times \chi=\chi/(1+z)$ where $\chi$ is the comoving distance.
    
A *comoving distance* between two objects in the Universe is the distance between them which remains constant with epoch if the two objects are moving with the Hubble flow (with the expansion of the Universe). Comoving distance factors out the expansion of the universe, giving a distance that does not change in time due to the expansion of space. The distance between two galaxy clusters is best considered in terms of a comoving distance.

The distance between stars in a galaxy, or the sizes of galaxies, are best considered in terms of a *physical* or *proper* distance.

    
By convention, the scale factor is set to 1 at the present time: $a(t_{today})=1$. As a result, the *comoving distance is equal to the proper distance at the present day*. This can also be seen by setting $z=0$ (today) in the equation    $r=a(t) \times \chi=\chi/(1+z)$

When computing proper and comoving distances will be different depending on if you are considering two points at the same redshift, or at different redshifts. At fixed redshift (e.g. the distance between two galaxies at the same redshift), one can use the equation given above.

    
Hubble parameter: $H(t)=\dot{a(t)}/a(t)$. The value of the Hubble parameter at the present day is called the Hubble parameter. It describes the present day expansion rate of the universe. $H_0 \sim 70$ km s$^{-1}$ Mpc$^{-1}$ (yes, these are odd units).

Critical Density: $\rho_{crit,0}=3H_0^2/8\pi G$

Density parameters: $\Omega_0=\rho_0/\rho_{crit,0}$, and $\Omega_0=\Omega_{m,0}+\Omega_{r_0}+\Omega_{\Lambda,0}$ (matter, radiation, and dark energy)
- non-relativisitic matter energy density changes as $\rho_m \propto a^{-3}$
- relativisitic matter (such as photons) energy density changes as $\rho_m \propto a^{-4}$ (the number density changes as $a^{-3}$ while the energy is redshifted according to $a^{-1}$)
- vacuum energy, of the cosmological cosntant $\Lambda$ whose density is a constant


The Friedmann-Robertson-Walker (FRW) Metric. The parameter K in the FRW metric is called the curvature parameter. It indicates the global curvature of space time. It can take on values +1,0,-1. Current constraints are consistent with a flat universe with K=0.

The *Friedmann equations* (named after <a href="https://en.wikipedia.org/wiki/Alexander_Friedmann">Alexander Friedmann</a>, who first derived them in 1922):

$$
\frac{\ddot{a}}{a}=-\frac{4\pi G}{3}\left(\rho+3\frac{P}{c^2}\right)+\frac{\Lambda c^2}{3}.
$$
$$
\left(\frac{\dot{a}}{a}\right)^2=\frac{8\pi G}{3}\rho+\frac{\Lambda c^2}{3}-\frac{Kc^2}{a^2}.
$$

The Friedmand equations relate the scale factor $a(t)$ to the energy density $\rho$ and the curvature $K$.

The more commonly used, compact form of the Friedmann equation is: $H(t)=H_0 \times E(z)$ where the $E(z)$ function is (assuming K=0 or $\Omega_0=1$):

$$E(z)=\sqrt{\Omega_{\Lambda,0}+\Omega_{M,0}(1+z)^3 + \Omega_{r,0}(1+z)^4}$$


*Notice the different redshift dependence for different components (dark energy, dark matter, radiation)*
